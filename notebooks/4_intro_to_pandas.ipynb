### Series in Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

s1 = pd.Series([1, 2, 3 , 4, 5])
type(s1)
```

```python
s2 = pd.Series(np.arange(5), dtype=np.uint8)
s2
```

```python
arr1 = np.arange(5, dtype=np.uint8)
s3 = pd.Series(arr1, dtype=np.int16)
s3
```

### Pandas Dataframes

```python
data = {'city': ['Nairobi', 'Nairobi', 'Nairobi',
 'Kampala', 'Kampala', 'Kampala'],
 'year': [2020, 2021, 2022, 2020, 2021, 2022,],
 'population': [10.0, 10.1, 10.2, 5.2, 5.3, 5.5]}

df1 = pd.DataFrame(data)
print(df1)
```

```python
df1.head()
```

```python
df2 = pd.DataFrame(data, columns=['year', 'city', 'population'])
print(df2)
```

### Visualizing the Data in Dataframes

```python
df1 = pd.DataFrame()
df1['A'] = pd.Series(list(range(100)))
df1['B'] = np.random.randn(100, 1)
df1

df1.plot(x='A', y='B')
plt.show()
```

```python
df2 = pd.DataFrame(np.random.rand(10, 4),
 columns=['A', 'B', 'C', 'D'])
print(df2)
```

```python
df2.plot.bar()
plt.show()
```

```python
df2.plot.barh()
plt.show()
```

```python
df2.plot.bar(stacked = True)
plt.show()
```

```python
df2.plot.hist(alpha=0.7)
plt.show()
```

```python
df2.plot.box()
plt.show()
```

```python
df2.plot.area()
plt.show()
```

```python
df2.plot.area(stacked=False)
plt.show()
```
