```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
y = x + 1
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(10)
y1 = 1 - x
plt.plot(x, y, x, y1)
plt.show()
```

```python
n = 5
x = np.arange(n)
y = np.random.rand(n)
plt.bar(x, y)
plt.show()
```

```python
x = np.arange(1, 10)

plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.title('Linear')

plt.subplot(2, 2, 2)
plt.plot(x, x*x)
plt.title('Quadratic')

plt.subplot(2, 2, 3)
plt.plot(x, np.sqrt(x))
plt.title('Square root')

plt.subplot(2, 2, 4)
plt.plot(x, np.log(x))
plt.title('Log')

plt.tight_layout()
plt.show()
```

```python
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
plt.scatter(x, y)
plt.show()
```

```python
mu, sigma = 0, 0.1
x = np.random.normal(mu, sigma, 1000)
plt.hist(x)
plt.show()
```

```python
x = np.array([10, 20, 30, 40])
plt.pie(x)
plt.show()
```
